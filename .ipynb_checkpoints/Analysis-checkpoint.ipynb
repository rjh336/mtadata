```python
import pandas as pd
data = pd.read_csv('./subway_csv')
data
```

```python
import re

# Downtown 6 at Brooklyn Bridge
sixtrain = data[(data['route_id'] == '6') & (data['stop_id'] == '640S')]

trip_id_int = [int(re.search(r"^[0-9]*", x).group(0)) for x in sixtrain['trip_id']]
sixtrain['trip_id_int'] = pd.Series(trip_id_int, index=sixtrain.index)
pd.to_numeric(sixtrain['arrival_time'], errors='coerce')
print sixtrain.trip_id.unique().size
print sixtrain['trip_id'].value_counts()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(sixtrain['trip_id_int'], sixtrain['arrival_time'])
```
